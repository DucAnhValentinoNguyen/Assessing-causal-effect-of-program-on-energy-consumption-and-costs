• Program “A” regions: Several regions implemented a program with strong financial incentives, including generous subsidies and tax credits for households undertaking comprehensive energy retrofits to their homes

• Program “B” regions: Two regions implemented a program that also aimed to reduce energy consumption but, due to budget limitations, used a different approach, relying primarily on information campaigns and small rebates rather than substantial financial incentives

• Control “C” regions: All other regions had no government programs incentivizing energy efficiency retrofits during this period. Households in control regions may still undertake retrofits, but without subsidies

Your task is therefore to produce credible estimates of
1. the causal effect of each program (Program A and Program B) on energy consump
tion and energy costs,

2. the causal effect of each program (Program A and Program B) on health,

3. the causal effect of retrofits themselves on energy consumption and energy costs.

4. investigate which type of households responded the most to the programs, i.e. for
which the causal effect of the programs on energy consumption and energy costs
was the smallest and largest, respectively
